## EMG

- FNSZ: Convulsões focais que não podem ser especificadas por seu tipo
- GNSZ: Crises generalizadas que não podem mais ser classificadas em um dos grupos abaixo
- SPSZ: Crises parciais durante a consciência; Tipo especificado apenas por sinais clínicos
- CPSZ: Convulsões parciais durante a inconsciência; Tipo especificado apenas por sinais clínicos
- ABSZ: Descargas de ausência observadas no EEG; paciente perde a consciência por alguns segundos (Petit Mal)
- TNSZ: Enrijecimento do corpo durante a convulsão (os efeitos do EEG desaparecem)
- CNSZ: Empurrão / tremor do corpo durante a convulsão
- TCSZ: No começo, enrijecendo e depois empurrando o corpo (Grand Mal)
- ATSZ: Perda repentina de tônus muscular
- MYSZ: Empurrões mioclônicos de membros
